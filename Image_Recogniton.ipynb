```python
import os
import random

folders = ['n02352591', 'n03908204', 'n07764155']

file_paths_with_class = []

class_names = {'n02352591': 'dormouse', 'n03908204': 'pencil', 'n07764155': 'mango'}


for folder in folders:
    files = os.listdir(folder)
    class_name = class_names.get(folder, 'unknown')
    file_paths_with_class.extend([(os.path.join(folder, file), class_name) for file in files])

random.shuffle(file_paths_with_class)

random_images = random.sample(file_paths_with_class, 10)

for image_path, class_name in random_images:
    image_id = os.path.basename(image_path).split('.')[0]
    print(f"Image: {image_id}, Class: {class_name}")


from PIL import Image
import matplotlib.pyplot as plt

for image_path, class_name in random_images:
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(f"Class: {class_name}")
    plt.axis('off')
    plt.show()
```
